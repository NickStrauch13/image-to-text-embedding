```python
import requests
import sqlite3

def is_valid_description(description):
    # Simple method to count sentences. This might need to be improved for accuracy.
    sentence_count = description.count('.') + description.count('!') + description.count('?')
    return 1 <= sentence_count <= 5

def fetch_and_save_artworks(api_url,connection):
    response = requests.get(api_url)
    if response.status_code != 200:
        print("Failed to fetch data")
        return
    count = 0
    data = response.json()
    # print(f"Fetched {len(data['data'])} artworks")
    for artwork in data['data']:
        if 'description' in artwork and artwork['description']:
            description = artwork['description']
            if is_valid_description(description):
                image_url = f"https://www.artic.edu/iiif/2/{artwork['image_id']}/full/843,/0/default.jpg"
                

                cursor = connection.cursor()
                # Insert the data into the table
                cursor.execute('INSERT INTO artworks VALUES (?, ?)', (image_url, description))
                # Commit the changes
                connection.commit()
                count += 1
    # print("Saved artworks to database")
    return count


connection = sqlite3.connect('artworks.db')
cursor = connection.cursor()
# Create a table if it doesn't exist
#delete the table if it exists
# cursor.execute('DROP TABLE IF EXISTS artworks')
cursor.execute('CREATE TABLE IF NOT EXISTS artworks (image_url TEXT, description TEXT)')
total_count = 0
for page in range(1000, 1500):
    # Create a connection to the database
    print(f"Fetching page {page}")
    api_url = f"https://api.artic.edu/api/v1/artworks?page={page}&limit=100"
    total_count+=fetch_and_save_artworks(api_url,connection)
    if total_count%100==0:
        print(f"Total artworks saved: {total_count}")
connection.close()
```
