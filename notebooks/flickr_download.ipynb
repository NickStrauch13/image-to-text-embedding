```python
import deeplake
import os
from PIL import Image
import numpy as np

# Load the dataset
ds = deeplake.load('hub://activeloop/flickr30k')

# Specify the directory where you want to save the images and captions
base_dir = "../data/flickr30k_images"
# os.makedirs(base_dir, exist_ok=True)

# Iterate through the dataset
for i, sample in enumerate(ds):
    if i >= 30000:  # Limit the number of images to download
        break

    # Convert the image from a numpy array to a PIL image and save it
    image = Image.fromarray(np.array(sample['image']))
    image_path = base_dir + f"/{i}.jpg"
    image.save(image_path)

    # Fetch each caption and convert it to string
    captions = [str(sample[f'caption_{j}'].numpy()) for j in range(5)]  # Convert tensor to numpy, then to string
    
    with open(os.path.join(base_dir, f"{i}_captions.txt"), "w") as caption_file:
        for caption in captions:
            caption_file.write(caption + "\n")

    print(f"Downloaded and saved image {i} and its captions.")

print("Download complete.")
```
